```python
# get Series Names
def clean_col_names(df):
    new_col_names = []
    for col in df.columns:
        if 'YR' in col:
            year, _ = col.split(' ')
            new_col_names.append(year)
        else:
            new_col_names.append(col.replace(' ', '_'))
    df.columns = new_col_names
    return df
```

```python
ed_data = clean_col_names(ed_data)
```
