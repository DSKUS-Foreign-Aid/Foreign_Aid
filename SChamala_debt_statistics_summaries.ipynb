```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp_graph_df = pd.read_csv('.\\data\\summary\\debt_population_gdp_latin_american.csv')
```

```python
gdp_graph_df.head(2)
```

## Graphs for Long Term Debt
- Long term debt to me tells that a country is borrowing at a lower interest
- Long term debt may also mean that the country has to pay more in interest because they pay over longer periods
- Long term debt may also mean that the country is not going be able to fulfill it's debts quickly
- Long term debt may also mean that the country is going to struggle in the short-term
- Long term debt may also mean that the country is trying to set a plan for the long-term

```python
plt.figure(figsize=(20,10))
plt.title('Amount Of Long Term Debt by Latin American Countries', fontsize = 20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Long Term Debt', fontsize=16)
debt_population_plot_df = gdp_graph_df[gdp_graph_df['debt_statistic'] =='Long-term external debt'].reset_index(drop=True)
sns.lineplot(x='year',y='value',data=debt_population_plot_df, hue='country')
plt.show()
```

```python
# The results show that Brazil, Mexico & Argentina having the most amount of Long-Term Debt
debt_population_plot_df[debt_population_plot_df['year']==2020].sort_values(by='value', ascending=False)[['country','value']].head(3)
```

```python
plt.figure(figsize=(20,10))
plt.title('Long-Term Debts Normalized By Population ', fontsize = 20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Long-term external debt Normalized by Population', fontsize=16)
debt_population_plot_df = gdp_graph_df[gdp_graph_df['debt_statistic'] =='Long-term external debt'].reset_index(drop=True)
sns.lineplot(x='year',y='debt_normalized_by_population',data=debt_population_plot_df, hue='country')
plt.show()
```

```python
# After normalizing by population the country that has the most amount of debt per person is Panama
debt_population_plot_df[debt_population_plot_df['year']==2020].sort_values(by='debt_normalized_by_population', ascending=False)[['country','debt_normalized_by_population']].head(3)
```

## Raw Long Term Debt vs Normalized Long Term Debt

- There are many debts that developing nations have This above graph shows the long-term debt some of the developing countries in Latin America. 
- The raw counts shows that Brazil, Mexico & Argentina have borrowed the most long-term debt. It is difficult though to make any conclusions just based on the raw numbers. So 1 option is to normalize this. The normalization that I did was by the population size for each of the respective years. 
- Based on the normaliation it can be seen that Panama has the highest debt compared to other developing nations in South America. 

```python
plt.figure(figsize=(20,10))
plt.title('Long-Term Debts Normalized By Population Vs. Log of GDP', fontsize = 20)
plt.xlabel('Long-term external debt Normalized by Population', fontsize=16)
plt.ylabel('Log of GDP', fontsize=16)
debt_population_plot_df = gdp_graph_df[gdp_graph_df['debt_statistic'] =='Long-term external debt'].reset_index(drop=True)
sns.scatterplot(x='debt_normalized_by_population', y='log_gdp',data=gdp_graph_df,hue='country')
plt.show()
```

## Scatter Plot of Long-term external debt(normalized by population) vs Log of GDP
- The scatter plot is showing if there exists a relationship between these 2 variables. None of the countries seemed to have a significant correlation. 
- Within each of the country's visualization there doesn't seem to be any specific relationship and it looks like most ofthe coutries stay within a range of GDP.
